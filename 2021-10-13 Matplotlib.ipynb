```python
x = [1, 2, 3]
y = [4, 9, 7]
```

```python
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
# 상태 기반 인터페이스 스타일
plt.plot(x, np.sin(x)); # ; 을 붙히게 되면 객체에 대한 설명이 사라지게 됨 
```

```python
ax = plt.axes()  # axis :  축, axes : 축들
ax.plot(x, np.sin(x)) # 객체지향 인터페이스 스타일
```

```python
plt.plot(x, np.sin(x), color='red', label='sin(x)') # color = '원하는 색상', label = 라벨링
plt.plot(x, np.cos(x), color='green', linestyle='--', label='cos(x)') # linestyle = '원하는 선의 형태'
plt.title('Sin & Cos Curves') # 제목
plt.xlabel('x') # 가로 축
plt.ylabel('f(y)') # 세로 축
plt.legend() # 범례, 그래프 구성에 대한 설명
plt.ylim((-1.2, 1.2)) # 축의 간격 늘림
plt.tight_layout() # 그래프만 나오게 함.
```

```python
np.random.seed(1)
z = np.random.randn(1000) # 표준정규분포 난수를 발생시킴
```

```python
plt.hist(z, bins=20) # bin : 구간, 칸, 깡통
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1.2, 1000)
```

```python
#plt.hist((x1, x2))의 방식으로 한번에 나타낼 수 있음
plt.hist(x1, alpha=0.7, bins=20) # 이 방식은 겹쳐지는 데이터 때문에 그래프가 가려지게 됨
plt.hist(x2, alpha=0.7, bins=20) # alpha는 투명도를 조절함
```
